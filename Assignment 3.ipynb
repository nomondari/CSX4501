Brief summary about data set: The Business Dynamics Statistics (BDS) includes measures of establishment openings and closings, firm startups, job creation and destruction by firm size, age, and industrial sector, and several other statistics on business dynamics. The U.S. economy is comprised of over 6 million establishments with paid employees. BDS is based on data going back through 1976. This allows business dynamics to be tracked, measured and analyzed for young firms in their first critical years as well as for more mature firms including those that are in the process of reinventing themselves in an ever changing economic environment.

Dataset Link: https://corgis-edu.github.io/corgis/python/business_dynamics/

```python
# Import the pandas library
import pandas as pd

businessdynamics_df = pd.read_csv('business_dynamics.csv')
```

```python
# Use pandas read_csv function to import the data into a dataframe variable
df = pd.DataFrame(businessdynamics_df)
```

```python
# Choose a column that has numeric values and make a line plot of the values
import matplotlib.pyplot as plt
plt.plot(businessdynamics_df['Data.Number of Firms'])
```

```python
# Use "loc" to print out the first 10 elements of the plotted column
df.iloc[:10].plot(kind='scatter', x='Data.Number of Firms', y='Year')
```

```python
# Use "loc" to print out the first 10 elements of the plotted column
# as well as the matching 10 elements of a different column that has interesting text

df.iloc[67:76].groupby('State').plot(kind='scatter', x='Data.Number of Firms', y='Year')
```

```python
# Assign the dataframe values from the previous cell into a new dataframe variable
# and make a bar plot with the text values horizontally and the numeric values as the bar heights
dfnew = pd.DataFrame(businessdynamics_df)

ax= dfnew.iloc[67:76].plot(kind='bar', 
                       x='Year', 
                       y='Data.Number of Firms', 
                       figsize=(14,5),
                       color=(0.5,0,0.5,0.25))
ax.legend(bbox_to_anchor=(1, 1))
ax.set_xlabel('Year',fontsize=10)
ax.set_ylabel('Number of Firms',fontsize=10)
```

```python
# Re-do the plot from the previous cell as a horizontal bar plot

ax= dfnew.iloc[67:76].plot(kind='barh', 
                           x='Year', 
                           y='Data.Number of Firms', 
                           figsize=(10,5),
                           color=(1,0,0.25,0.25))
ax.legend(bbox_to_anchor=(1, 1))
ax.set_ylabel('Year',fontsize=10)
ax.set_xlabel('Number of Firms',fontsize=10)
```

```python
# Re-do the plot from the previous cell
# and change at least two aesthetic elements (colors, labels, titles, ...)

ax= dfnew.iloc[67:76].plot(kind='barh', 
                           x='Year', 
                           y='Data.Number of Firms', 
                           figsize=(10,5),
                           color=(1,0,0.25,0.25))
ax.legend(bbox_to_anchor=(1, 1))
ax.set_title('Number of Firms in Alaska')
ax.set_ylabel('Year',fontsize=10)
ax.set_xlabel('Number of Firms',fontsize=10)
ax.figure.savefig('BarhAlaska.png')
```
